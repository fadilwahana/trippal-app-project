Data 2

```python
data2 = data2[['Jenis Kelamin', 'Umur', 'Pekerjaan', 'Hobi', 'Preferensi Perjalanan 1', 'Preferensi Perjalanan 2', 'Preferensi Perjalanan 3']]
```

```python
from sklearn.preprocessing import LabelEncoder

#Encode Jenis Kelamin Data 2
le = LabelEncoder()
var_to_encode = data2['Jenis Kelamin']
encoded_var = le.fit_transform(var_to_encode)
data2['Jenis Kelamin'] = encoded_var

#Encode Hobi Data 2
var_to_encode = data2['Hobi']
encoded_var = le.fit_transform(var_to_encode)
data2['Hobi'] = encoded_var

#Encode Preferensi Perjalanan 1 Data 2
var_to_encode = data2['Preferensi Perjalanan 1']
encoded_var = le.fit_transform(var_to_encode)
data2['Preferensi Perjalanan 1'] = encoded_var

#Encode Preferensi Perjalanan 2 Data 2
var_to_encode = data2['Preferensi Perjalanan 2']
encoded_var = le.fit_transform(var_to_encode)
data2['Preferensi Perjalanan 2'] = encoded_var

#Encode Preferensi Perjalanan 3 Data 2
var_to_encode = data2['Preferensi Perjalanan 3']
encoded_var = le.fit_transform(var_to_encode)
data2['Preferensi Perjalanan 3'] = encoded_var

#Encode Pekerjaan Data 2
var_to_encode = data2['Pekerjaan']
encoded_var = le.fit_transform(var_to_encode)
data2['Pekerjaan'] = encoded_var
```

**TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
data1_train, data1_test, data2_train, data2_test = train_test_split(data1, data2, test_size=0.2, random_state=1, shuffle=True)
```

```python
print(data1_train.shape)
print(data1_test.shape)
print(data2_train.shape)
print(data2_test.shape)
```

Neural Network - Content Based Filtering

```python
num_data1_features = data1_train.shape[1]
num_data2_features = data2_train.shape[1]
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
num_outputs = 32
tf.random.set_seed(1)
data1_NN = tf.keras.models.Sequential([ 
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_outputs, activation='linear'),
])

data2_NN = tf.keras.models.Sequential([
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_outputs, activation='linear'),
])
```

```python
#data1 = user, data2 = item

# create the user input and point to the base network
input_data1 = tf.keras.layers.Input(shape=(num_data1_features))
vu = data1_NN(input_data1)
vu = tf.linalg.l2_normalize(vu, axis=1)

# create the item input and point to the base network
input_data2 = tf.keras.layers.Input(shape=(num_data2_features))
vm = data2_NN(input_data2)
vm = tf.linalg.l2_normalize(vm, axis=1)

# compute the dot product of the two vectors vu and vm
output = tf.keras.layers.Dot(axes=1)([vu, vm])

# specify the inputs and output of the model
model = tf.keras.Model([input_data1, input_data2], output)

model.summary()
```

Mean Square Error Loss and Adam Oprimizer

```python
tf.random.set_seed(1)
cost_fn = tf.keras.losses.MeanSquaredError()
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,
              loss=cost_fn,metrics=['accuracy'])
```

```python
uvs = 0
ivs = 0
u_s = 0
i_s = 0
```

```python
tf.random.set_seed(1)

# Convert DataFrames to NumPy arrays
data1_train_array = data1_train.iloc[:, u_s:].values
data2_train_array = data2_train.iloc[:, i_s:].values

# Replicate the target value to match the number of samples
target_data_array = np.repeat(1, data1_train_array.shape[0])

history = model.fit(
    [data1_train_array, data2_train_array],
    target_data_array,
    epochs=100,
    verbose=1
)
```

```python
print("Input data shape:", data1_train_array.shape)
print("Target data shape:", target_data_array.shape)
```

```python
model.evaluate([data1_test.iloc[:, u_s:], data2_test.iloc[:, i_s:]])
```

Plot of Accuracy and Loss

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
#val_acc = history.history['val_accuracy']

loss = history.history['loss']
#val_loss = history.history['val_loss']

epochs_range = np.arange(100)

plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
#plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
#plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
