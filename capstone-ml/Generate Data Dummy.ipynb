```python
from faker import Faker
import random
import pandas as pd
```

```python
faker = Faker('id_ID')
```

```python
records = []
gender_list = ['Pria', 'Wanita']
for i in range(20000):
  gender = random.choice(gender_list)
  if gender == 'Pria':
    name = faker.name_male() 
  else:
    name = faker.name_female() 
  
  email_name = name.replace(" ", "").replace(",", "").replace(".", "").lower()
  email = f"{email_name}@example.com"
  dob = faker.date_of_birth(minimum_age = 17, maximum_age = 50)
  address = faker.address()
  city = faker.city()
  job = faker.job()
  phone = f"08{str(faker.random_int(min=10000000000, max=99999999999))}"
  hobby_list = ['Membaca', 'Menulis', 'Bersepeda', 'Fotografi', 'Memasak', 'Berkebun', 'Mendaki', 'Menyelam']
  hobby = random.choice(hobby_list)
  preference_list = ['Wisata Kuliner', 'Wisata Alam', 'Wisata Seni dan Budaya', 'Wisata Sejarah dan Religi']
  preference = random.choice(preference_list)
  sim_list = ['-','SIM A', 'SIM C', 'SIM A dan C']
  sim = random.choice(sim_list)

  records.append({'Nama' : name, 'Email': email, 'Jenis Kelamin': gender, 'Tanggal Lahir' : dob, 'Kota Domisili' : city, 'Pekerjaan' : job, 'No HP' : phone, 'Hobi' : hobby, 'Preferensi Perjalanan' : preference, 'SIM' : sim})
```

```python
df = pd.DataFrame(records)
```
